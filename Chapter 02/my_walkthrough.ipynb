## 2D drawing in Python

A set of **custom-built** functions contained in **vector_drawing.py** are used to create drawings, built on top of _another_ Python library **Matplotlib**. The following Python classes represent drawable geometric objects:

1. class Polygon()
2. class Points()
3. class Arrow()
4. class Segment()

In the example above, points (6,4) and (3,1) are connected by the **draw** function and the **Points** and **Segment** classes which are an abstraction for drawing a line between two points. 


```python
draw(
    Points(*dino_vectors),
    Polygon(*dino_vectors)
)
```

**Description**: 

Run the **draw** function on the Points class on vector (x, x**2) for x in the range from x = -10 to x = 11. **grid** is a parameter in the draw function that is by default set to (1,1), but can be adjusted here as (1,10) and the **nice_aspect_ratio** which tells the graph it doesn't have to keep x-axis and y-axis scales the same, set to False. 

```python
draw(
    Points(*[(x,x**2) for x in range(-10,11)]),
    grid=(1,10),
    nice_aspect_ratio=False
)
```

# Plane Vector Arithmetic


```python
def add(v1, v2):
    return (v1[0] + v2[0], v1[1] + v2[1])
```

```python
dino_vectors2 = [add((-1.5, -2.5), v) for v in dino_vectors]
```

```python
draw(
    Points(*dino_vectors, color=blue),
    Polygon(*dino_vectors, color=blue),
    Points(*dino_vectors2, color=red),
    Polygon(*dino_vectors2, color=red)
)
```

```python
arrows = [Segment(tip, tail, color=black) for (tip,tail) in
         zip(dino_vectors2, dino_vectors)]

draw(
    Points(*dino_vectors, color=blue),
    Polygon(*dino_vectors, color=blue),
    Points(*dino_vectors2, color=red),
    Polygon(*dino_vectors2, color=red),
    *arrows
)
```
